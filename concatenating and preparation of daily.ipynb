```python
full_data2 = full_data.copy()
full_data2["Confirmed_cum"] = full_data.groupby("Country/Region")["Confirmed"].cumsum()
full_data2["Deaths_cum"] = full_data.groupby("Country/Region")["Deaths"].cumsum()
full_data2["Recovered_cum"] = full_data.groupby("Country/Region")["Recovered"].cumsum()
```

```python
full_data2["activos"] = full_data2["Confirmed_cum"] -full_data2["Deaths_cum"] - full_data2["Recovered_cum"]
```
